#**Horizontal Barh Graph of 'Location'**


```python
#Horizontal Barh Graph of 'Location'
plt.figure(figsize = (20, 10))
ax = df["Location"].value_counts().plot(kind = 'barh',color = sns.color_palette('tab20'))
```

#N**o. of Occurrences of 'Gender' By using Bar Graph**


```python
#No. of Occurrences of 'Gender' By using Bar Graph
plt.figure(figsize=(20,6))
a = (df.Gender.value_counts()).plot(kind='bar',color = sns.color_palette('tab20'))
plt.xlabel("Gender",weight="bold",color='red')
plt.ylabel('Number of Occurrences', weight = "bold", color = "red")
a.legend()
plt.show()
```

#**Bar plot for 'Categoty'**


```python
#Bar plot for 'Categoty'
plt.figure(figsize=(10,4))
(df.Category.value_counts()).plot(kind='bar',label='Category',color = sns.color_palette('tab20'),width=0.2)
plt.xlabel('Category')
plt.ylabel('No. of Occurrences')
plt.legend()
plt.show()
```

#**Pie Chart for 'Gender'**


```python
#Pie Chart for 'Gender'
a=(df.Gender.value_counts()).plot(kind ='pie',fontsize = 12, autopct ='%.f%%',explode=[0,0.1])
a.legend()
```

#**'Item_Puchased' plot by barh graph**


```python
#'Item_Puchased' plot by barh graph
plt.figure(figsize = (25, 8))
df.Item_Purchased.value_counts().sort_values(ascending = True).plot(kind = 'barh', color = sns.color_palette('tab20'), edgecolor = 'black')
plt.ylabel('Item_Purchased', fontsize = 16)
plt.xlabel('\nNumber of Occurrences', fontsize = 16)
plt.title('Item Purchased\n', fontsize = 16)
plt.xticks(rotation = 0, fontsize = 16)
plt.tight_layout()
plt.legend()
plt.show()
```

# **'Frequency_of_Purchase' Pie Chart **

```python
plt.figure(figsize = (20, 6))

counts = df["Frequency_of_Purchases"].value_counts()
explode = (0, 0, 0, 0, 0.0, 0, 0.06)

counts.plot(kind = 'pie', fontsize = 12, explode = (0, 0, 0, 0, 0.0, 0, 0.06), autopct = '%2.1f%%')
plt.xlabel('Quarter', weight = "bold", color = "#2F0F5D", fontsize = 14, labelpad = 20)
plt.axis('equal')
plt.legend(labels = counts.index, loc = "best")
plt.show()
```

#**What is the maximum and minimum review rating in the dataset ?**

```python
maximum_rating = df["Review_Rating"].max()
minimum_rating = df["Review_Rating"].min()
print('Maximum Rating given by Customer : ',maximum_rating)
print('Minimum Rating given by Customer : ',minimum_rating)
```

#**What is the total purchase amount for customers in each location ?**


```python
total_purchase_by_location = df.groupby('Location')["Purchase_Amount_(USD)"].sum()
print("Total Purchase Amount by Location : ",total_purchase_by_location)
```

#**KDE plot for Purchase Amount by Season**


```python
plt.figure(figsize=(20,8))
sns.kdeplot(data = df, x = 'Purchase_Amount_(USD)', hue = 'Season', common_norm = False, fill = True, palette = 'Set1')
plt.title('KDE Plot for Purchase Amount by Season')
plt.xlabel('Purchase_Amount_(USD)')
plt.show()
```

#**Scatter Plot of Age vs. Review Rating**

```python
plt.figure(figsize=(20,6))
plt.scatter(x = df['Age'], y = df['Review_Rating'], alpha=0.5)
plt.title('Age vd Review Rating')
plt.xlabel('Age')
plt.ylabel('Review_Rating')
plt.show()
```

#**Bar Plot by Payment Method By Season**

```python
plt.figure(figsize=(20,6))
sns.barplot(data = df,x=df['Season'],y=df['Purchase_Amount_(USD)'], hue ='Payment_Method')
plt.title('Bar Plot By Payment Method By Season')
plt.xlabel('Season')
plt.ylabel('Purchase_Amount_(USD)')
plt.show()
```

#**Histogram By Age Distribution**


```python
plt.figure(figsize=(20,6))
plt.hist(df['Age'],bins=20,edgecolor='k')
plt.title('AGE Distribution')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# PIE chart by Payment Methods

plt.figure(figsize = (20, 6))
counts = df["Payment_Method"].value_counts()
counts.plot(kind = 'pie', fontsize = 12,explode = (0, 0, 0, 0, 0.0, 0.06), autopct = '%1.1f%%')
plt.xlabel('Size', weight = "bold", color = "#2F0F5D", fontsize = 14, labelpad = 20)
plt.axis('equal')
plt.legend(labels = counts.index, loc = "best")
plt.show()
```
